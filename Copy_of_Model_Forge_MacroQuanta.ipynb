```python
# ==========================================================
# INDIA + GLOBAL GOLD‚ÄìSILVER MACRO DATA COLLECTION
# ==========================================================

import yfinance as yf
import pandas as pd
import datetime
import os
import time
from pandas_datareader import data as pdr


# ------------------------------
# 1Ô∏è‚É£ DATE RANGE
# ------------------------------
START_DATE = "2005-01-01"
END_DATE = datetime.datetime.today()


# ------------------------------
# 2Ô∏è‚É£ CREATE FOLDER STRUCTURE
# ------------------------------
BASE_PATH = "data1"
INDIA_PATH = os.path.join(BASE_PATH, "india_data")
GLOBAL_PATH = os.path.join(BASE_PATH, "global_data")

os.makedirs(INDIA_PATH, exist_ok=True)
os.makedirs(GLOBAL_PATH, exist_ok=True)


# ==========================================================
# 3Ô∏è‚É£ YAHOO DOWNLOAD FUNCTION
# ==========================================================

def download_yahoo(ticker, name, save_path):
    try:
        df = yf.download(
            ticker,
            start=START_DATE,
            end=END_DATE,
            auto_adjust=False,
            progress=False
        )

        if df.empty:
            print(f"{name} not available.")
            return

        if "Adj Close" in df.columns:
            df = df[["Adj Close"]].rename(columns={"Adj Close": name})
        else:
            df = df[["Close"]].rename(columns={"Close": name})

        df.reset_index(inplace=True)
        df.to_csv(os.path.join(save_path, f"{name}.csv"), index=False)

        print(f"Saved: {name}")

    except Exception as e:
        print(f"Error downloading {name}: {e}")


# ==========================================================
# 4Ô∏è‚É£ GLOBAL MARKET DATA (Yahoo)
# ==========================================================

global_tickers = {
    # Precious Metals Futures
    "Gold_Futures": "GC=F",
    "Silver_Futures": "SI=F",

    # USD & FX
    "DXY": "DX-Y.NYB",
    "EURUSD": "EURUSD=X",
    "USDJPY": "JPY=X",

    # Risk & Equities
    "SP500": "^GSPC",
    "VIX": "^VIX",

    # Commodities
    "Copper": "HG=F",
    "Oil": "CL=F",

    # Volatility Structure Proxy
    "VIX3M": "^VIX3M"
}

print("\nDownloading GLOBAL market data...\n")

for name, ticker in global_tickers.items():
    download_yahoo(ticker, name, GLOBAL_PATH)
    time.sleep(0.3)


# ==========================================================
# 5Ô∏è‚É£ INDIA MARKET DATA (Yahoo)
# ==========================================================

india_tickers = {
    "Gold_India_ETF": "GOLDBEES.NS",
    "Silver_India_ETF": "SILVERBEES.NS",
    "USDINR": "USDINR=X",
    "NIFTY50": "^NSEI",
    "India_VIX": "^INDIAVIX",

    # Banking & Financial Stress Proxy
    "BANKNIFTY": "^NSEBANK",

    # Oil (India imports heavy oil ‚Üí affects INR & inflation)
    "Brent_Oil": "BZ=F"
}

print("\nDownloading INDIA market data...\n")

for name, ticker in india_tickers.items():
    download_yahoo(ticker, name, INDIA_PATH)
    time.sleep(0.3)


# ==========================================================
# 6Ô∏è‚É£ FRED MACRO DATA
# ==========================================================

def download_fred(series_code, name, save_path):
    try:
        df = pdr.DataReader(series_code, "fred", START_DATE, END_DATE)

        if df.empty:
            print(f"{name} not available from FRED.")
            return

        df.reset_index(inplace=True)
        df.rename(columns={series_code: name}, inplace=True)
        df.to_csv(os.path.join(save_path, f"{name}.csv"), index=False)

        print(f"Saved: {name}")

    except Exception as e:
        print(f"Error downloading {name} from FRED")


fred_series = {
    # ------------------
    # US Real Rate Drivers
    # ------------------
    "US_Real_Yield_10Y": "DFII10",
    "US_10Y_Yield": "DGS10",
    "US_2Y_Yield": "DGS2",

    # Inflation Expectations
    "US_Breakeven_10Y": "T10YIE",
    "US_Breakeven_5Y": "T5YIE",
    "US_CPI": "CPIAUCSL",

    # Liquidity
    "US_M2": "M2SL",

    # Growth
    "US_Industrial_Production": "INDPRO",

    # ------------------
    # INDIA MACRO
    # ------------------
    # -----------------------
    # üü° Inflation
    # -----------------------
    "India_CPI_All": "INDCPIALLMINMEI",

    # -----------------------
    # üü° Interest Rates
    # -----------------------
    "India_Policy_Rate": "INTDSRINM193N",
    "India_10Y_Yield": "INDIRLTLT01STM",
    "India_3M_TBill": "INDIR3TIB01STM",

    # -----------------------
    # üü° Liquidity & Money Supply
    # -----------------------
    "India_M3": "MYAGM3INM189N",

    # -----------------------
    # üü° Growth & Production
    # -----------------------
    "India_GDP_Per_Capita": "PCAGDPINA646NWDB",

    # -----------------------
    # üü° Banking & Financial Conditions
    # -----------------------
    "India_Bank_NPL_Ratio": "DDOI11INA156NWDB",
}

print("\nDownloading FRED macro data...\n")

for name, code in fred_series.items():
    # US macro saved in global folder
    if name.startswith("US_"):
        download_fred(code, name, GLOBAL_PATH)
    else:
        download_fred(code, name, INDIA_PATH)

    time.sleep(0.3)


print("\n‚úÖ INDIA + GLOBAL GOLD‚ÄìSILVER DATA COLLECTION COMPLETE.")
```

Starting with the first step of the plan: Data Preparation. I will load the collected data from the `data1` directory, merge them, convert to business-day frequency, forward-fill macro variables, set `Date` as the time index, and ensure no missing critical asset values.

```python
import pandas as pd
import os

# Define the paths for India and Global data
INDIA_PATH = "data1/india_data"
GLOBAL_PATH = "data1/global_data"

def load_and_merge_data(path, df=None):
    """Loads CSV files from a given path and merges them into a single DataFrame."""
    for filename in os.listdir(path):
        if filename.endswith('.csv'):
            filepath = os.path.join(path, filename)
            try:
                temp_df = pd.read_csv(filepath)
                if 'Date' not in temp_df.columns:
                    print(f"Skipping {filename}: 'Date' column not found.")
                    continue

                temp_df['Date'] = pd.to_datetime(temp_df['Date'])
                temp_df = temp_df.set_index('Date').sort_index()

                if df is None:
                    df = temp_df
                else:
                    df = df.merge(temp_df, on='Date', how='outer')
                print(f"Loaded and merged: {filename.replace('.csv', '')}")
            except Exception as e:
                print(f"Error loading {filename}: {e}")
    return df

# Initialize an empty DataFrame
combined_df = None

# Load and merge global data
combined_df = load_and_merge_data(GLOBAL_PATH, combined_df)

# Load and merge India data
combined_df = load_and_merge_data(INDIA_PATH, combined_df)

# Ensure Date is the index and sort it
combined_df = combined_df.sort_index()

print(f"\nInitial combined DataFrame shape: {combined_df.shape}")
print("First 5 rows of the combined DataFrame:")
display(combined_df.head())
```

### Subtask:

Finally, I will display the head, tail, and shape of the `features_df` to verify the output.



```python
import numpy as np

# Identify all numeric columns in combined_df (all columns except the 'Date' index)
numeric_cols = combined_df.select_dtypes(include=np.number).columns.tolist()

# Initialize a new DataFrame for features with the same index as combined_df
features_df = pd.DataFrame(index=combined_df.index)

# Loop through each identified numeric column to generate features
for col in numeric_cols:
    # Calculate Log Return
    features_df[f'{col}_LogReturn'] = np.log(combined_df[col] / combined_df[col].shift(1))

    # Calculate 20-Day Rolling Volatility (of Log Return)
    features_df[f'{col}_Volatility20D'] = features_df[f'{col}_LogReturn'].rolling(window=20).std()

    # Calculate 20-Day Rolling Mean (Momentum) (of original column's values)
    features_df[f'{col}_Momentum20D'] = combined_df[col].rolling(window=20).mean()

print(f"\nInitial combined_df shape: {combined_df.shape}")
print(f"Generated features_df shape: {features_df.shape}")

print("\nFirst 5 rows of features_df:")
display(features_df.head())

print("\nLast 5 rows of features_df:")
display(features_df.tail())
```

**Reasoning**:
The `RuntimeWarning`s (`invalid value encountered in log`, `divide by zero encountered in log`) indicate that `np.log` was called with non-positive arguments or a division by zero occurred. To prevent these warnings and ensure robust log return calculations, I will modify the log return calculation to explicitly handle non-positive ratios or zero denominators by coercing them to `NaN` before applying the logarithm. This ensures `np.log` only receives positive arguments.



```python
import numpy as np

# Identify all numeric columns in combined_df (all columns except the 'Date' index)
numeric_cols = combined_df.select_dtypes(include=np.number).columns.tolist()

# Initialize a new DataFrame for features with the same index as combined_df
features_df = pd.DataFrame(index=combined_df.index)

# Loop through each identified numeric column to generate features
for col in numeric_cols:
    # Calculate Log Return
    # Ensure previous_day_value is not zero and the ratio is positive before taking log
    previous_day_values = combined_df[col].shift(1)

    # Replace 0 in previous_day_values with NaN to prevent division by zero warning
    # Then calculate the ratio. If ratio is not positive, it will be converted to NaN before log.
    ratio = combined_df[col] / previous_day_values.replace(0, np.nan)
    features_df[f'{col}_LogReturn'] = np.log(ratio.where(ratio > 0))

    # Calculate 20-Day Rolling Volatility (of Log Return)
    features_df[f'{col}_Volatility20D'] = features_df[f'{col}_LogReturn'].rolling(window=20).std()

    # Calculate 20-Day Rolling Mean (Momentum) (of original column's values)
    features_df[f'{col}_Momentum20D'] = combined_df[col].rolling(window=20).mean()

print(f"\nInitial combined_df shape: {combined_df.shape}")
print(f"Generated features_df shape: {features_df.shape}")

print("\nFirst 5 rows of features_df:")
display(features_df.head())

print("\nLast 5 rows of features_df:")
display(features_df.tail())
```

### Explanation of Walk-Forward Training Methodology

**1. Walk-Forward Training Approach for Time-Series Forecasting:**
Walk-forward training is a robust model validation technique specifically designed for time-series data. Unlike a fixed train-test split, which assumes that the data distribution remains constant over time, walk-forward training simulates real-world trading or forecasting scenarios. It involves chronologically moving through the dataset, training a model on a segment of historical data (the "training window"), and then making predictions for a subsequent, immediate future period (the "test period"). This process is repeated sequentially, with the training window advancing forward in time, continuously updating the model's knowledge as new data becomes available. This methodology is crucial for time-series forecasting because financial markets and economic conditions are dynamic and evolve, making past relationships potentially irrelevant or misleading for future predictions if models are not regularly updated.

**2. Rolling 252-Day Training Window:**
For our analysis, we will employ a **rolling training window of 252 business days**. This period roughly corresponds to one year of trading days, providing a balance between capturing recent market dynamics and having sufficient data for robust model training. Each time a prediction is made for a new test observation, the model will be trained on the most recent 252 days of available data. For example:
*   To predict for `Day T`, the model will be trained on data from `Day T-252` to `Day T-1`.
*   To predict for `Day T+1`, the training window will slide forward one day, using data from `Day T-251` to `Day T`.
This continuous advancement ensures that the model is always leveraging the most current market information, allowing it to react to recent trends and changes.

**3. Retraining Models Before Each Test Prediction:**
Before generating each new prediction for a single test day (or a short test period), the chosen models‚ÄîRandom Forest Regressor, Gradient Boosting Regressor, and Ridge Regression‚Äîwill be fully retrained. This means that for every forward step in the test set:
*   The features and target variables corresponding to the current 252-day training window will be extracted.
*   Each model will be re-initialized and fitted from scratch using this updated training data.
*   Once trained, the model will be used to make a prediction for the next out-of-sample observation (e.g., the next single day).
This intensive retraining schedule is vital because it ensures that the model's parameters and learned relationships are continuously adapted to the latest market conditions, providing the most up-to-date predictive capability.

**4. Prevention of Look-Ahead Bias:**
The walk-forward training methodology inherently prevents **look-ahead bias** because, at any given point in time, the model is only exposed to data that would have been historically available. When the model is trained for a specific prediction date, it only uses data *prior* to that date. It never uses future information to make current predictions. This strict chronological separation ensures that the evaluation of the model's performance is a fair and realistic representation of how it would perform in a live forecasting environment.

**5. Handling Regime Shifts:**
The combination of a rolling training window and frequent retraining provides a powerful mechanism for handling **regime shifts**. Financial markets often experience periods with distinct characteristics (e.g., high volatility, low volatility, bull markets, bear markets, economic expansions, recessions). A model trained on data from a past regime might perform poorly in a new, different regime. By continuously retraining on a rolling window of recent data, the models can:
*   **Adapt Quickly**: As the market transitions into a new regime, the rolling window will gradually include more data from the new regime and shed older data from the previous one. This allows the model to learn the new underlying patterns.
*   **Forget Obsolete Information**: Older data, which might be less relevant to the current market environment, is progressively dropped from the training set, preventing the model from being overly influenced by outdated relationships. This dynamic adjustment enables the models to remain responsive and maintain predictive power even when market conditions undergo significant changes.

## Define Evaluation Metrics

### Subtask:
Specify the evaluation metrics to be used, differentiating between Forecasting Metrics (RMSE, MAE, R¬≤, Directional Accuracy, multi-horizon performance) and Trading Metrics (Sharpe Ratio, CAGR, Maximum Drawdown, Sortino Ratio, Hit Ratio), along with their respective weights.


### Evaluation Metrics Specification

To comprehensively evaluate the performance of our models for predicting Gold and Silver Futures, we will consider two main categories of metrics: Forecasting Metrics and Trading Metrics. These metrics will allow us to assess both the accuracy of our predictions and the profitability and risk of strategies derived from these predictions.

#### 1. Forecasting Metrics
These metrics will primarily focus on the accuracy and reliability of the model's predictions for future price movements. We will evaluate multi-horizon performance for the 1-day, 30-day, and 126-day forward log returns.

*   **a. Root Mean Squared Error (RMSE)**:
    *   **Description**: Measures the average magnitude of the errors. It is the square root of the average of the squared differences between prediction and actual observation. RMSE gives a relatively high weight to large errors, meaning it is most useful when large errors are particularly undesirable.

*   **b. Mean Absolute Error (MAE)**:
    *   **Description**: Measures the average magnitude of the errors in a set of predictions, without considering their direction. MAE is less sensitive to outliers compared to RMSE, as it uses absolute differences rather than squared differences.

*   **c. R-squared (R¬≤)**:
    *   **Description**: Represents the proportion of the variance in the dependent variable that is predictable from the independent variables. In time-series forecasting, a positive R¬≤ indicates that the model explains some of the variance in the target. It's often difficult to achieve high R¬≤ in financial time series.

*   **d. Directional Accuracy (DA)**:
    *   **Description**: Measures the percentage of times the model correctly predicts the direction of the price movement (up or down). This is a crucial metric for trading strategies, as correctly predicting the direction is often more important than the exact magnitude of the change.

*   **e. Multi-horizon Performance Evaluation**:
    *   **Description**: We will assess RMSE, MAE, R¬≤, and Directional Accuracy separately for each defined forecast horizon (1-day, 30-day, and 126-day). This ensures that the model's performance is robust across different timeframes relevant to our trading and investment objectives.

#### 2. Trading Metrics
These metrics will assess the financial viability and risk-adjusted returns of a hypothetical trading strategy based on the model's predictions.

*   **a. Sharpe Ratio**:
    *   **Description**: Measures the excess return (or risk premium) per unit of total risk in an investment asset or a trading strategy. A higher Sharpe Ratio indicates a better risk-adjusted return.

*   **b. Compound Annual Growth Rate (CAGR)**:
    *   **Description**: Represents the geometric mean annual growth rate of an investment over a specified time period. It smooths out volatile returns and provides a more accurate representation of annual growth than a simple arithmetic mean.

*   **c. Maximum Drawdown (MDD)**:
    *   **Description**: The maximum observed loss from a peak to a trough of a portfolio, before a new peak is attained. MDD is an indicator of downside risk over a specified time period.

*   **d. Sortino Ratio**:
    *   **Description**: Similar to the Sharpe Ratio, but it only penalizes downside volatility. It measures the risk-adjusted return of an asset or strategy relative to its negative deviations, providing a more focused view on harmful volatility.

*   **e. Hit Ratio**:

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# --- 1. Define helper functions and calculate Forecasting Metrics (1-day ahead) ---

def calculate_directional_accuracy(y_true, y_pred, threshold=0.0):
    """
    Calculates the directional accuracy of predictions.
    Considers predictions within +/- threshold as 'no change'.
    """
    true_direction = np.sign(y_true)
    pred_direction = np.sign(y_pred)

    # Handle cases where actual or predicted values are very close to zero
    # If actual is zero, we don't count it towards a correct direction unless predicted is also zero
    # If predicted is zero, it's correct if actual is also zero, otherwise it's incorrect

    # Ensure we only compare non-zero actuals or where both are zero
    correct_predictions = (true_direction == pred_direction) & (np.abs(y_true) > threshold)
    # If actual is very close to zero, and predicted is also very close to zero, count as correct
    correct_predictions = correct_predictions | ((np.abs(y_true) <= threshold) & (np.abs(y_pred) <= threshold))

    return np.sum(correct_predictions) / len(y_true) if len(y_true) > 0 else 0

print("\n--- Forecasting Metrics (1-day ahead) ---")

print("\nGold Futures:")
# RMSE
gold_rmse = np.sqrt(mean_squared_error(actual_gold_series, predicted_gold_series))
print(f"  RMSE: {gold_rmse:.6f}")

# MAE
gold_mae = mean_absolute_error(actual_gold_series, predicted_gold_series)
print(f"  MAE: {gold_mae:.6f}")

# R2
gold_r2 = r2_score(actual_gold_series, predicted_gold_series)
print(f"  R-squared: {gold_r2:.4f}")

# Directional Accuracy
gold_da = calculate_directional_accuracy(actual_gold_series, predicted_gold_series, threshold=1e-5)
print(f"  Directional Accuracy: {gold_da:.2%}")

print("\nSilver Futures:")
# RMSE
silver_rmse = np.sqrt(mean_squared_error(actual_silver_series, predicted_silver_series))
print(f"  RMSE: {silver_rmse:.6f}")

# MAE
silver_mae = mean_absolute_error(actual_silver_series, predicted_silver_series)
print(f"  MAE: {silver_mae:.6f}")

# R2
silver_r2 = r2_score(actual_silver_series, predicted_silver_series)
print(f"  R-squared: {silver_r2:.4f}")

# Directional Accuracy
silver_da = calculate_directional_accuracy(actual_silver_series, predicted_silver_series, threshold=1e-5)
print(f"  Directional Accuracy: {silver_da:.2%}")
```

```python
import pandas as pd
import numpy as np
import os
import datetime
from pandas_datareader import data as pdr
import yfinance as yf
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Ridge

# --- Re-execute Data Preparation, Feature Engineering, and Time-Series Split --- #
# This section is re-included to ensure 'final_features_df' and 'df_test' are defined.

# 1. Date Range & Folder Structure (from _EkaitDHUvv8 & 2‚Ä∞ CREATE FOLDER STRUCTURE)
START_DATE = "2005-01-01"
END_DATE = datetime.datetime.today()
BASE_PATH = "data1"
INDIA_PATH = os.path.join(BASE_PATH, "india_data")
GLOBAL_PATH = os.path.join(BASE_PATH, "global_data")

# Ensure folders exist
os.makedirs(INDIA_PATH, exist_ok=True)
os.makedirs(GLOBAL_PATH, exist_ok=True)

# ==========================================================
# 3‚Ä∞ YAHOO DOWNLOAD FUNCTION (Copied from _EkaitDHUvv8)
# ==========================================================

def download_yahoo(ticker, name, save_path):
    try:
        df = yf.download(
            ticker,
            start=START_DATE,
            end=END_DATE,
            auto_adjust=False,
            progress=False
        )

        if df.empty:
            print(f"{name} not available.")
            return

        if "Adj Close" in df.columns:
            df = df[["Adj Close"]].rename(columns={"Adj Close": name})
        else:
            df = df[["Close"]].rename(columns={"Close": name})

        df.reset_index(inplace=True)
        df.to_csv(os.path.join(save_path, f"{name}.csv"), index=False)

        # print(f"Saved: {name}") # Suppress output for re-run

    except Exception as e:
        print(f"Error downloading {name}: {e}")


# ==========================================================
# 4‚Ä∞ GLOBAL MARKET DATA (Yahoo) (Copied from _EkaitDHUvv8)
# ==========================================================

global_tickers = {
    # Precious Metals Futures
    "Gold_Futures": "GC=F",
    "Silver_Futures": "SI=F",

    # USD & FX
    "DXY": "DX-Y.NYB",
    "EURUSD": "EURUSD=X",
    "USDJPY": "JPY=X",

    # Risk & Equities
    "SP500": "^GSPC",
    "VIX": "^VIX",

    # Commodities
    "Copper": "HG=F",
    "Oil": "CL=F",

    # Volatility Structure Proxy
    "VIX3M": "^VIX3M"
}

# print("\nDownloading GLOBAL market data...\n") # Suppress output for re-run

for name, ticker in global_tickers.items():
    download_yahoo(ticker, name, GLOBAL_PATH)
    # time.sleep(0.3) # Avoid sleep in this self-contained block


# ==========================================================
# 5‚Ä∞ INDIA MARKET DATA (Yahoo) (Copied from _EkaitDHUvv8)
# ==========================================================

india_tickers = {
    "Gold_India_ETF": "GOLDBEES.NS",
    "Silver_India_ETF": "SILVERBEES.NS",
    "USDINR": "USDINR=X",
    "NIFTY50": "^NSEI",
    "India_VIX": "^INDIAVIX",

    # Banking & Financial Stress Proxy
    "BANKNIFTY": "^NSEBANK",

    # Oil (India imports heavy oil ‚Üí affects INR & inflation)
    "Brent_Oil": "BZ=F"
}

# print("\nDownloading INDIA market data...\n") # Suppress output for re-run

for name, ticker in india_tickers.items():
    download_yahoo(ticker, name, INDIA_PATH)
    # time.sleep(0.3) # Avoid sleep in this self-contained block


# ==========================================================
# 6‚Ä∞ FRED MACRO DATA (Copied and modified from _EkaitDHUvv8 to match b7a2abb9)
# ==========================================================

def download_fred(series_code, name, save_path):
    try:
        df = pdr.DataReader(series_code, "fred", START_DATE, END_DATE)

        if df.empty:
            print(f"{name} not available from FRED.")
            return

        df.reset_index(inplace=True)
        df.rename(columns={'DATE': 'Date'}, inplace=True) # Rename the date column
        df.rename(columns={series_code: name}, inplace=True)
        df[name] = pd.to_numeric(df[name], errors='coerce') # Convert data column to numeric
        df.to_csv(os.path.join(save_path, f"{name}.csv"), index=False)

        # print(f"Saved: {name}") # Suppress output for re-run

    except Exception as e:
        print(f"Error downloading {name} from FRED")


fred_series = {
    # ------------------
    # US Real Rate Drivers
    # ------------------
    "US_Real_Yield_10Y": "DFII10",
    "US_10Y_Yield": "DGS10",
    "US_2Y_Yield": "DGS2",

    # Inflation Expectations
    "US_Breakeven_10Y": "T10YIE",
    "US_Breakeven_5Y": "T5YIE",
    "US_CPI": "CPIAUCSL",

    # Liquidity
    "US_M2": "M2SL",

    # Growth
    "US_Industrial_Production": "INDPRO",

    # ------------------
    # INDIA MACRO
    # ------------------
    # ‚óè Inflation
    "India_CPI_All": "INDCPIALLMINMEI",

    # ‚óè Interest Rates
    "India_Policy_Rate": "INTDSRINM193N",
    "India_10Y_Yield": "INDIRLTLT01STM",
    "India_3M_TBill": "INDIR3TIB01STM",

    # ‚óè Liquidity & Money Supply
    "India_M3": "MYAGM3INM189N",

    # ‚óè Growth & Production
    "India_GDP_Per_Capita": "PCAGDPINA646NWDB",

    # ‚óè Banking & Financial Conditions
    "India_Bank_NPL_Ratio": "DDOI11INA156NWDB",
}

# print("\nDownloading FRED macro data...\n") # Suppress output for re-run

for name, code in fred_series.items():
    # US macro saved in global folder
    if name.startswith("US_"):
        download_fred(code, name, GLOBAL_PATH)
    else:
        download_fred(code, name, INDIA_PATH)

    # time.sleep(0.3) # Avoid sleep in this self-contained block


# print("\n‚úÖ INDIA + GLOBAL GOLD‚ÄìSILVER DATA COLLECTION COMPLETE.\n") # Suppress output for re-run


# 2. load_and_merge_data function (from 683883f9, modified)
def load_and_merge_data(path, df=None):
    """Loads CSV files from a given path and merges them into a single DataFrame."""
    # print(f"Attempting to load data from: {path}") # Suppress verbose output once fixed
    files_in_dir = os.listdir(path)
    if not files_in_dir:
        print(f"No CSV files found in directory: {path}")
        return df

    for filename in files_in_dir:
        if filename.endswith('.csv'):
            filepath = os.path.join(path, filename)
            try:
                # Attempt to read, explicitly setting 'Date' as index and parsing dates
                temp_df = pd.read_csv(filepath, parse_dates=['Date'], index_col='Date')
                temp_df = temp_df.sort_index() # Ensure index is sorted

                # Convert all columns (excluding the new Date index) to numeric, coercing errors
                for col in temp_df.columns:
                    temp_df[col] = pd.to_numeric(temp_df[col], errors='coerce')

                if df is None:
                    df = temp_df
                else:
                    # Merge on the index (Date)
                    df = df.merge(temp_df, left_index=True, right_index=True, how='outer')
                # print(f"Successfully loaded and merged: {filename.replace('.csv', '')}") # Suppress output for re-run
            except KeyError:
                # This error occurs if 'Date' column is not found to set as index_col
                print(f"Skipping {filename}: 'Date' column not found or incorrectly named. Check FRED/Yahoo download output.")
            except Exception as e:
                print(f"Error processing {filename}: {e}")
    return df

# Initialize an empty DataFrame
combined_df = None

# Load and merge global data
combined_df = load_and_merge_data(GLOBAL_PATH, combined_df)

# Load and merge India data
combined_df = load_and_merge_data(INDIA_PATH, combined_df)

# Ensure Date is the index and sort it (this line should now be safe if combined_df is not None)
if combined_df is not None:
    combined_df = combined_df.sort_index()
else:
    raise ValueError("No data loaded after attempting to merge from all paths. combined_df is None.")

# 3. Convert to business-day frequency and forward-fill (from 9b21e892)
combined_df_b = combined_df.asfreq('B')
combined_df_b_filled = combined_df_b.ffill()
combined_df = combined_df_b_filled # Update combined_df for next steps

# 4. Feature Engineering (from e01b58b7 & b4f4e077)
numeric_cols = combined_df.select_dtypes(include=np.number).columns.tolist()
features_df = pd.DataFrame(index=combined_df.index)

for col in numeric_cols:
    previous_day_values = combined_df[col].shift(1)
    ratio = combined_df[col] / previous_day_values.replace(0, np.nan)
    features_df[f'{col}_LogReturn'] = np.log(ratio.where(ratio > 0))
    features_df[f'{col}_Volatility20D'] = features_df[f'{col}_LogReturn'].rolling(window=20).std()
    features_df[f'{col}_Momentum20D'] = combined_df[col].rolling(window=20).mean()

final_features_df = pd.concat([combined_df, features_df], axis=1)

# 5. Target Creation (from 73ecfeff & 50edd8c6)
future_gold_price = final_features_df['Gold_Futures'].shift(-1)
final_features_df['Gold_Futures_Target_LogReturn'] = np.log(future_gold_price / final_features_df['Gold_Futures'])

future_silver_price = final_features_df['Silver_Futures'].shift(-1)
final_features_df['Silver_Futures_Target_LogReturn'] = np.log(future_silver_price / final_features_df['Silver_Futures'])

future_gold_price_30d = final_features_df['Gold_Futures'].shift(-30)
final_features_df['Gold_Futures_Target_LogReturn_30D'] = np.log(future_gold_price_30d / final_features_df['Gold_Futures'])

future_silver_price_30d = final_features_df['Silver_Futures'].shift(-30)
final_features_df['Silver_Futures_Target_LogReturn_30D'] = np.log(future_silver_price_30d / final_features_df['Silver_Futures'])

future_gold_price_126d = final_features_df['Gold_Futures'].shift(-126)
final_features_df['Gold_Futures_Target_LogReturn_126D'] = np.log(future_gold_price_126d / final_features_df['Gold_Futures'])

future_silver_price_126d = final_features_df['Silver_Futures'].shift(-126)
final_features_df['Silver_Futures_Target_LogReturn_126D'] = np.log(future_silver_price_126d / final_features_df['Silver_Futures'])

# 6. Time-Series Split (from b8759db6)
train_start = '2005-01-01'
train_end = '2018-12-31'
val_start = '2019-01-01'
val_end = '2021-12-31'
test_start = '2022-01-01'
test_end = '2025-12-31'

df_train = final_features_df.loc[train_start:train_end]
df_val = final_features_df.loc[val_start:val_end]
df_test = final_features_df.loc[test_start:test_end]

# 7. Handle NaNs in target columns (from 2b02df97)
target_columns = [
    'Gold_Futures_Target_LogReturn',
    'Silver_Futures_Target_LogReturn',
    'Gold_Futures_Target_LogReturn_30D',
    'Silver_Futures_Target_LogReturn_30D',
    'Gold_Futures_Target_LogReturn_126D',
    'Silver_Futures_Target_LogReturn_126D'
]

df_train = df_train.dropna(subset=target_columns)
df_val = df_val.dropna(subset=target_columns)
df_test = df_test.dropna(subset=target_columns)

# --- End of Re-execution of Prerequisites --- #


# Instantiate the models with placeholder parameters (re-instantiate to ensure fresh models for retraining)
rfr = RandomForestRegressor(random_state=42, n_estimators=100, max_depth=10, n_jobs=-1)
gbr = GradientBoostingRegressor(random_state=42, n_estimators=100, max_depth=5)
ridge = Ridge(alpha=1.0)

# Store models in a list for easy iteration
ensemble_models = [
    ('RandomForest', rfr),
    ('GradientBoosting', gbr),
    ('Ridge', ridge)
]

# 1. Define the 1-day ahead target variable columns
target_cols_1d_ahead = ['Gold_Futures_Target_LogReturn', 'Silver_Futures_Target_LogReturn']
all_target_columns = [
    'Gold_Futures_Target_LogReturn',
    'Silver_Futures_Target_LogReturn',
    'Gold_Futures_Target_LogReturn_30D',
    'Silver_Futures_Target_LogReturn_30D',
    'Gold_Futures_Target_LogReturn_126D',
    'Silver_Futures_Target_LogReturn_126D'
]

# 2. Define the feature columns
feature_columns = [col for col in final_features_df.columns if col not in all_target_columns]

# 3. Initialize empty lists to store predictions, actuals, and dates
gold_predictions = []
silver_predictions = []
gold_actuals = []
silver_actuals = []
prediction_dates = []

# 4. Set the rolling window size
rolling_window_size = 252 # business days

# Find the index of the first date in df_test within the full dataset
start_idx_test = final_features_df.index.get_loc(df_test.index[0])

# Create a copy of the models to ensure they are re-initialized for each walk-forward step
def get_fresh_models():
    return [
        ('RandomForest', RandomForestRegressor(random_state=42, n_estimators=100, max_depth=10, n_jobs=-1)),
        ('GradientBoosting', GradientBoostingRegressor(random_state=42, n_estimators=100, max_depth=5)),
        ('Ridge', Ridge(alpha=1.0))
    ]

for i in range(start_idx_test, len(final_features_df)):
    current_test_date = final_features_df.index[i]

    # Only process dates that are actually in our df_test range
    if current_test_date not in df_test.index:
        continue

    # Ensure we have enough historical data for the rolling window
    if i < rolling_window_size:
        continue

    # Slice final_features_df for the training window
    current_training_data_full_range = final_features_df.iloc[max(0, i - rolling_window_size):i]

    # Drop any rows from current_training_data that contain NaN values in any of the feature columns or the 1-day target columns
    current_training_data = current_training_data_full_range.dropna(subset=feature_columns + target_cols_1d_ahead)

    if current_training_data.empty:
        continue

    # Separate X_train and y_train_gold, y_train_silver
    X_train = current_training_data[feature_columns]
    y_train_gold = current_training_data['Gold_Futures_Target_LogReturn']
    y_train_silver = current_training_data['Silver_Futures_Target_LogReturn']

    # Extract X_test_day for the current_test_date
    X_test_day = final_features_df.loc[[current_test_date], feature_columns]

    # If X_test_day contains any NaN values in its feature columns, skip prediction
    if X_test_day.isnull().any().any():
        continue

    # Re-initialize models for each step to ensure no state carries over
    current_ensemble_models = get_fresh_models()

    individual_gold_preds = []
    individual_silver_preds = []

    # Train and predict for each model in the ensemble
    for name, model in current_ensemble_models:
        # Gold Prediction
        model.fit(X_train, y_train_gold)
        gold_pred = model.predict(X_test_day)[0]
        individual_gold_preds.append(gold_pred)

        # Silver Prediction
        model.fit(X_train, y_train_silver)
        silver_pred = model.predict(X_test_day)[0]
        individual_silver_preds.append(silver_pred)

    # Calculate the ensemble average prediction
    ensemble_gold_pred = np.mean(individual_gold_preds)
    ensemble_silver_pred = np.mean(individual_silver_preds)

    # Store the ensemble predictions and actuals
    gold_predictions.append(ensemble_gold_pred)
    silver_predictions.append(ensemble_silver_pred)
    gold_actuals.append(df_test.loc[current_test_date, 'Gold_Futures_Target_LogReturn'])
    silver_actuals.append(df_test.loc[current_test_date, 'Silver_Futures_Target_LogReturn'])
    prediction_dates.append(current_test_date)

# Convert collected lists into pandas Series
predicted_gold_series = pd.Series(gold_predictions, index=prediction_dates, name='Gold_Predicted')
predicted_silver_series = pd.Series(silver_predictions, index=prediction_dates, name='Silver_Predicted')
actual_gold_series = pd.Series(gold_actuals, index=prediction_dates, name='Gold_Actual')
actual_silver_series = pd.Series(silver_actuals, index=prediction_dates, name='Silver_Actual')

# Print the total number of predictions generated
print(f"Total 1-day ahead predictions generated: {len(predicted_gold_series)}")

# Display the head of the prediction and actual series
print("\nGold Predictions and Actuals (Head):")
display(pd.DataFrame({'Predicted': predicted_gold_series, 'Actual': actual_gold_series}).head())

print("\nSilver Predictions and Actuals (Head):")
display(pd.DataFrame({'Predicted': predicted_silver_series, 'Actual': actual_silver_series}).head())
```
